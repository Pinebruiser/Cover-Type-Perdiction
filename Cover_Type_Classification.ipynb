**Checking for Outliers**

```python
data[['Elevation','Slope','Aspect']].plot(kind='box')
plt.show()
```

```python
columns_to_check = ['Elevation', 'Aspect', 'Slope']

for col in columns_to_check:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_bound) | (data[col] > upper_bound)][col]

    print(f"Outliers for {col}:")
    display(outliers)
```

**too many outliers to drop

**Building the model**

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler,LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,GridSearchCV
```

**Encoding because the cover type is 1,2,3..etc while the model expects 0,1,2..etc**

```python
le = LabelEncoder()
data['Cover_Type'] = le.fit_transform(data['Cover_Type'])
```

**Splitting the data**

```python
X=data.drop('Cover_Type',axis=1)
y= data['Cover_Type']
```

```python
from sklearn.model_selection import train_test_split

# temp train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
y_train.value_counts()
```

```python
xgb_params={
    'n_jobs':-1,
    'random_state':10,
    'n_estimators':1000,
    'learning_rate':0.05
}
```

```python
XGB= XGBClassifier(**xgb_params)
RFC= RandomForestClassifier(random_state=10)
```

**testing and training scores for XGB**

```python
from sklearn.metrics import confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
XGB.fit(X_train,y_train)
```

```python
XGB.score(X_test,y_test)
```

```python
XGB.score(X_train,y_train)
```

```python
y_pred = XGB.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
cm =confusion_matrix(y_test,y_pred,labels=XGB.classes_)
disp= ConfusionMatrixDisplay(cm,display_labels=XGB.classes_)
disp.plot(cmap='BuPu_r')
plt.show()
```

```python
importances = XGB.feature_importances_
feature_names = X_train.columns
forest_importances = pd.Series(importances, index=feature_names)

fig, ax = plt.subplots(figsize=(10, 8))
forest_importances.sort_values().plot.barh(ax=ax)
ax.set_title("Feature importances using XGB Classifier")
ax.set_ylabel("Mean decrease in impurity")
fig.tight_layout()
plt.show()
```

**testing and training scores for Random Forest**

```python
RFC.fit(X_train,y_train)
```

```python
RFC.score(X_test,y_test)
```

```python
RFC.score(X_train,y_train)
```
